```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20 ,25 ,30 ,40]
```

```python
plt.plot(x, y);
plt.title('Simple line plot');
plt.xlabel('X-axis label');
plt.ylabel('Y-axis label');
plt.grid(True);
```

```python
month = ['Jan', 'Feb', 'Mar', 'Apr']
revenue = [100, 200, 300, 500]
```

```python
plt.figure(figsize=(10, 6));
plt.bar(month, revenue);
plt.title('Monthly revenue bar chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
plt.grid(True);
```

```python
heights = [65, 70, 68, 72, 66]
weights = [150, 160, 155, 180, 145]
```

```python
plt.scatter(heights, weights)
plt.xlabel('Height in inches')
plt.ylabel('Weights in pounds')
```

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
predicted_price = [8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o', label='Original price');
plt.plot(days, predicted_price, 'r-o', label='Predicted price');
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('Price')
```
